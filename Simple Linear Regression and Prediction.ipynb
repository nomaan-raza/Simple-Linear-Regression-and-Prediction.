## What is Simple Linear Regression?

In statistics, simple linear regression is a linear regression model with a single explanatory variable.
In simple linear regression, we predict scores on one variable based on results on another.

The criteria variable Y is the variable we are predicting.
Predictor variable X is the variable using which we are making our predictions.

The prediction approach is known as simple regression as there is only one predictor variable.

Simple Linear Regression in machine learning with Python is a method for modeling the relationship between a dependent variable and a single independent variable by fitting a linear equation to observed data.

Linear Regression-

Linear regression uses the relationship between the data-points to draw a straight line through all them.
This line can be used to predict future values.

```python
import pandas as pd

# Create the dataframe (replace with your actual data if needed)

df = pd.DataFrame({'Weight': [60, 70, 80, 90], 'Height': [160, 170, 180, 190]})

X = df[['Weight']]
y = df[['Height']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
## Standardization

from sklearn.preprocessing import StandardScaler
```

```python
## Using Scaler as Standardization

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
## Scaler to test

X_test = scaler.transform(X_test)
```

```python
# Printing test output

X_test
```

```python
## Apply Simple Linear Regression

from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression(n_jobs=-1)
```

```python
## LinearRegression fit

regression.fit(X_train,y_train)
```

```python
# Printing Coefficient or slope and Intercept values

print("Coefficient or slope:",regression.coef_)
print("Intercept:",regression.intercept_)
```

```python
import matplotlib.pyplot as plt

##plot training data plot best fit line
## Plotting scatter type data graph line
## graph title as Training Data and Best Fit Line

plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train), color='red')
plt.xlabel('Standardized Weight')
plt.ylabel('Height')
plt.title('Training Data and Best Fit Line')
plt.show()
```

Prediction of test data
1. Predicted height output=intercept+coef_(Weights)
2. y_pred_test=156.470+17.29(X_test)

```python
## Prediction for test data

y_data=regression.predict(X_test)
```

```python
## Importing and Performance Metrics
## from sklearn module importing mean_absolute_error,mean_squared_error.

from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
import numpy as np

# Flatten y_test to 1D arrays for metric calculation
# Flatten y_data to 1D arrays for metric calculation

mse = mean_squared_error(y_test.values.ravel(), y_data.ravel())
mae = mean_absolute_error(y_test.values.ravel(), y_data.ravel())
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

R Square

Formula
R^2=1-SSR/SST
R^2= coefficient of determination SSR=sum of squares of residuals SSt=total sum of squares

```python
# Importing R Square from sklearn module

from sklearn.metrics import r2_score
```

```python
# Making Prediction

y_pred=regression.predict(X_test)
```

```python
# Defining key value for r2_score

score=r2_score(y_test,y_pred)
```

```python
# Printing the Mean Squared Error value
# Printing the R-Squared value

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {score}")
```

```python
# Using 
import matplotlib.pyplot as plt
# Plotting graph with scatter type graph line
# Labelling the x and y axis as Independent Variable (X) and Dependent Variable (y)
# Giving graph name as Simple Linear Regression

plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X,regression.predict(X), color='red', label='Regression line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
